```python
lstCity = dataframe['Shipping City'].unique()
```

```python
def cityFilter(x):
    city = x.split(' ')
    if(len(city) > 1):
        return city[0]
    return 0
```

```python
lstMainCity = []
for i in lstCity:
    ct = cityFilter(i)
    if  ct != 0:
        lstMainCity.append(ct)
```

```python
lstMainCity
```

```python
def clearSpaceProgramCity(x):
    city = x.split(' ')
    cityChange = False
    if len(city) > 1:
        for i in city:
            for j in lstMainCity:
                if i == j:
                    city = j
                    cityChange = True
                    break
    if cityChange ==True:
        return city
    else:
        strCity = ''
        for ele in city: 
            strCity += ele + ' '
        return strCity
```

```python
dataframe['Shipping City'] = dataframe['Shipping City'].apply(clearSpaceProgramCity)
dataframe['Shipping City'] = dataframe['Shipping City'].apply(lambda x: x.strip())
dataframe['Shipping City'] =dataframe['Shipping City'].str.replace("chennai-600113","chennai")
dataframe['Shipping City'] =dataframe['Shipping City'].str.replace("delhi.110092","delhi")
```
