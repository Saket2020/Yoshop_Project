```python
import pandas as pd
import numpy as np


name_person = 'Name'
phone_no = 'Contact Number'
food_item_no = 'Food Item No '
b_name = 'Biryani Name'
Quantity_biryani = 'Quantity'

def billing_system(name, contact, Food_id, biryani_name, qty, cost):
    print('-' * 60)
    print('\t\t\t YOSHOPS AMBUR BIRYANI \t\t\t\t')
    print('-' * 60)
    print(f"{name_person.upper()}\t\t\t\t{name}")
    print(' ' * 60)
    print(f"{phone_no.upper()}\t\t\t\t{contact}")
    print(' ' * 60)
    print(f"{food_item_no.upper()}\t\t\t\t{Food_id}")
    print(' ' * 60)
    print(f"{b_name.upper()}\t\t\t\t{biryani_name}")
    print(' ' * 60)
    print(f"{Quantity_biryani.upper()}\t\t\t\t{qty}")
    print(' ' * 60)
    print('-' * 60)
    print(f'TOTAL \t\t\t\t{cost}')
    print('\n')

name = input('Enter the name of the customer: ')
contact = int(input('Enter the customer contact number: '))

total_cost = 0  # Variable to store the total bill amount

while True:
    Food_id = int(input('Enter the food id (Enter 0 to stop): '))
    if Food_id == 0:
        break

    if Food_id == 1:
        biryani_name = 'Chicken Biryani'
        cost = 170.00
    elif Food_id == 2:
        biryani_name = 'Chicken Biryani'
        cost = 120.00
    elif Food_id == 3:
        biryani_name = 'Chicken Biryani'
        cost = 80.00
    elif Food_id == 4:
        biryani_name = 'Kushka'
        cost = 110.00
    elif Food_id == 5:
        biryani_name = 'Kushka'
        cost = 70.00
    elif Food_id == 6:
        biryani_name = 'Kushka'
        cost = 50.00
    elif Food_id == 7:
        biryani_name = 'Egg Biryani'
        cost = 120.00
    elif Food_id == 8:
        biryani_name = 'Egg Biryani'
        cost = 80.00
    elif Food_id == 9:
        biryani_name = 'Mutton Biryani'
        cost = 260.00
    elif Food_id == 10:
        biryani_name = 'Mutton Biryani'
        cost = 190.00
    else:
        print('Invalid ID')
        continue

    qty = int(input('Enter the quantity: '))
    item_cost = cost * qty
    total_cost += item_cost

    billing_system(name, contact, Food_id, biryani_name, str(qty) + ' kg', item_cost)

print(f'Total Bill Amount: {total_cost}')
```
